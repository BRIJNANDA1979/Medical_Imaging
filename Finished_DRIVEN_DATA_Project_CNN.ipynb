```python
                                #####   2.1   Create Training Data
# category = ['antelope_duiker', 'bird', 'blank', 'civet_genet', 'hog', 'leopard', 'monkey_prosimian', 'rodent']
```

```python
category = ['antelope_duiker', 'bird', 'blank', 'civet_genet', 'hog', 'leopard', 'monkey_prosimian', 'rodent']
```

```python
import os
import cv2
```

```python

training = []
IMG_SIZE = 200
training_data_path = '/content/drive/MyDrive/DRIVENDATA_for_Submission/TRAIN_8_Categoried_Data'
def createTrainingData():
  for cat in category:
    path = os.path.join(training_data_path, cat)
    class_num =category.index(cat)               ######   [0,1,2,3,4,5,6,7]
    for img in os.listdir(path):
      print(img)
      img_array = cv2.imread(os.path.join(path,img))
      if img_array is None:
        print('Image file not found or corrupted.')
      else:
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        training.append([new_array, class_num])


```

```python
createTrainingData()          ### time Consuming task
```

```python
import random                                     #############  Shuffle the Dataset
random.shuffle(training)
```

```python
                              ############ label data
```

```python
training[0]
```

```python
import numpy as np
X =[]
y =[]
for features, label in training:
  X.append(features)
  y.append(label)
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 3)    # reshape to remove RGB i.e. 3 from (200,200,3) to get (200,200) shape
```

```python
len(X)
```

```python
len(y)
```

```python
for i in range(len(y)):
  if y[i]==0:
    y[i] = [1,0,0,0,0,0,0,0]
  if y[i]==1:
    y[i] = [0,1,0,0,0,0,0,0]
  if y[i]==2:
    y[i] = [0,0,1,0,0,0,0,0]
  if y[i]==3:
    y[i] = [0,0,0,1,0,0,0,0]
  if y[i]==4:
    y[i] = [0,0,0,0,1,0,0,0]
  if y[i]==5:
    y[i] = [0,0,0,0,0,1,0,0]
  if y[i]==6:
    y[i] = [0,0,0,0,0,0,1,0]
  if y[i]==7:
    y[i] = [0,0,0,0,0,0,0,1]


```

```python
y=np.array(y)
```

```python
X = X.astype('float32')
X /= 255
```

```python
import sklearn.model_selection
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.2, random_state = 4)
```

```python
len(X_test)
```

```python
import tensorflow as tf
```

```python
batch_size = 16
nb_classes =8
nb_epochs = 1
img_rows, img_columns = 200, 200
img_channel = 3
nb_filters = 32
nb_pool = 2
nb_conv = 3

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), padding='same', activation=tf.nn.relu,
                           input_shape=(200, 200, 3)),
    tf.keras.layers.MaxPooling2D((2, 2), strides=2),
    tf.keras.layers.Conv2D(32, (3,3), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=2),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(8,  activation=tf.nn.softmax)
])
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=[tf.keras.metrics.CategoricalAccuracy(),
                       tf.keras.metrics.FalseNegatives()])
model.fit(X_train, np.array(y_train), batch_size = batch_size, epochs = nb_epochs, verbose = 1, validation_data = (X_test, y_test))
```

```python
score = model.evaluate(X_test, y_test, verbose = 0 )
print("Test Score: ", score[0])
print("Test accuracy: ", score[1])
```
